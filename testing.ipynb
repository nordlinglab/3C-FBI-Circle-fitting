```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
# !pip install numba

from CCC_FBI import ccc_fbi
from preprocessing import preprocess_image, get_all_method_names
print("Imports complete:")
print(f'Available preprocessing methods: {get_all_method_names()}')
```

```python
image_id = 0
bs_img = cv2.imread(BS[image_id])
gb_img = cv2.imread(GB[image_id])

# Display images
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(cv2.cvtColor(bs_img, cv2.COLOR_BGR2RGB))
axes[0].set_title('Black Sphere ROI')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(gb_img, cv2.COLOR_BGR2RGB))
axes[1].set_title('Green Background ROI')
axes[1].axis('off')
plt.tight_layout()
plt.show()

print(f"Black Sphere image shape: {bs_img.shape}")
print(f"Green Background image shape: {gb_img.shape}")
```

```python
# Cell 3: Test single preprocessing method
method = 'GL76'  # Try: GL70, GL72, ..., GL86 or Med3, Med5, ..., Med19

# Preprocess to get edgels
if method.startswith('Med'):
    edgels = preprocess_image(bs_img, gb_img, method=method, return_edgels=True)
else:
    edgels = preprocess_image(bs_img, method=method, return_edgels=True)

if edgels is False:
    print(f"❌ No edges detected with method {method}")
else:
    print(f"✓ Detected {len(edgels)} edge points")
    print(f"Edgels shape: {edgels.shape}")
```

```python
# Cell 4: Visualize edgels
if edgels is not False:
    # Get edge image for visualization
    if method.startswith('Med'):
        edge_img = preprocess_image(bs_img, gb_img, method=method, return_edgels=False)
    else:
        edge_img = preprocess_image(bs_img, method=method, return_edgels=False)
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Original image
    axes[0].imshow(cv2.cvtColor(bs_img, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Original Black Sphere')
    axes[0].axis('off')
    
    # Edge image
    axes[1].imshow(edge_img, cmap='gray')
    axes[1].set_title(f'Edges ({method})')
    axes[1].axis('off')
    
    # Edgel points overlay
    axes[2].imshow(cv2.cvtColor(bs_img, cv2.COLOR_BGR2RGB))
    axes[2].scatter(edgels[:, 0], edgels[:, 1], c='red', s=16)
    axes[2].set_title(f'Edgels: {len(edgels)} points')
    axes[2].axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("No edgels to visualize")
```

```python
# Cell 5: Fit circle using 3C-FBI
if edgels is not False and len(edgels) >= 3:
    # Fit circle
    center, radius = ccc_fbi(edgels, Nmax=1000, xmax=100, ymax=100, rmin=5, rmax=50)
    
    if center[0] != -1:
        print(f"✓ Circle detected!")
        print(f"  Center: ({center[0]:.2f}, {center[1]:.2f})")
        print(f"  Radius: {radius:.2f}")
    else:
        print("❌ Circle fitting failed")
else:
    print("❌ Insufficient edgels for circle fitting")
```
