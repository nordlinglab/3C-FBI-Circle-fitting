```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from pathlib import Path
from PIL import Image

# ── Options ──────────────────────────────────────────────────────────────────
N_SHOW = 12          # number of images to display (randomly picked)
# ─────────────────────────────────────────────────────────────────────────────

DATA_DIR = Path('data')
IMG_DIR, EXT = DATA_DIR / 'black_sphere_ROI', '.png'

gt = pd.read_csv(DATA_DIR / 'Ground_Truth.csv')
gt.head()
```

```python
rows = gt.sample(n=N_SHOW, random_state=None)
cols = 3
n_rows = (len(rows) + cols - 1) // cols

fig, axes = plt.subplots(n_rows, cols, figsize=(5 * cols, 5 * n_rows))
axes = axes.flatten()

for ax, (_, row) in zip(axes, rows.iterrows()):
    img_path = IMG_DIR / (row['Filename'] + EXT)
    img = Image.open(img_path)
    ax.imshow(img, cmap='gray' if img.mode == 'L' else None)
    circle = patches.Circle(
        (row['X'], row['Y']), row['R'],
        linewidth=2, edgecolor='red', facecolor='none'
    )
    ax.add_patch(circle)
    ax.set_title(row['Filename'].split('_')[-2] + '_' + row['Filename'].split('_')[-1], fontsize=9)
    ax.axis('off')

for ax in axes[len(rows):]:
    ax.set_visible(False)

plt.suptitle("Ground Truth Circles — black_sphere_ROI", fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```
